```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_r = 80
n_z = 161
r = np.linspace(0.01, 1.0, n_r)
z = np.linspace(-1.0, 1.0, n_z)
[r_grid, z_grid] = np.meshgrid(r, z)
r_flat = r_grid.flatten()
z_flat = z_grid.flatten()

psi_b = -0.059069372604888676
xpt_r = 0.33894642418735055
xpt_z = -0.5713182906388565
xpt_psi_b = -0.0590836619459338

mag_r_previous = 0.5115792196972574
mag_z_previous = -0.007343976139471093
vessel_r = np.loadtxt("vessel_r.txt")
vessel_z = np.loadtxt("vessel_z.txt")

psi2d = np.loadtxt("psi2d.txt")
psi_2d = psi2d.reshape((161, 80))

mask_2d = np.loadtxt("mask_2d.txt").reshape((161, 80))
```

```python
plt.figure(figsize=(10, 8))
plt.axis("equal")
plt.contour(r, z, psi_2d)
plt.plot(r_flat, z_flat, "k.", markersize=1)
plt.plot(vessel_r, vessel_z, color="red", marker="o", linewidth=2)
plt.contour(r, z, psi_2d, levels=[psi_b], colors="b", linewidths=2)
plt.contour(r, z, psi_2d, levels=[xpt_psi_b], colors="r", linewidths=2)
plt.xlim(0, 0.5)
plt.ylim(-0.7, -0.5)
plt.plot(xpt_r, xpt_z, color="red", marker="x", markersize=10)

index_inside_plasma = mask_2d == 1
index_outside_plasma = mask_2d == 0

plt.plot(r_grid[index_inside_plasma], z_grid[index_inside_plasma], color="green", marker="o", linestyle="none", markersize=2)
plt.plot(r_grid[index_outside_plasma], z_grid[index_outside_plasma], color="black", marker="o", linestyle="none", markersize=0.5)
```
