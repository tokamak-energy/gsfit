```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import mu_0
```

```python
%matplotlib inline
```

```python

# Given global center coordinates
grid_centers_r = np.linspace(0.01, 1.01, 81)
grid_centers_z = np.linspace(-1.0, 1.0, 161)

# Define a conductor with current
conductor_center_r = grid_centers_r[32]
conductor_center_z = grid_centers_z[80]

grid_centers_downsampled_r = grid_centers_r[10:50]
grid_centers_downsampled_z = grid_centers_z[60:100]

# Cell size from spacing
d_r = grid_centers_downsampled_r[1] - grid_centers_downsampled_r[0]
d_z = grid_centers_downsampled_z[1] - grid_centers_downsampled_z[0]

# Meshgrid of cell centres
r_grid, z_grid = np.meshgrid(grid_centers_downsampled_r, grid_centers_downsampled_z)
r_flat = r_grid.flatten()
z_flat = z_grid.flatten()

# Plot
plt.figure(figsize=(6.0, 6.0))
plt.axis('equal')

for i in range(len(r_flat)):
    if r_flat[i] == conductor_center_r and z_flat[i] == conductor_center_z:
        rect = plt.Rectangle(
            (r_flat[i] - d_r / 2, z_flat[i] - d_z / 2),
            d_r, d_z,
            facecolor='#ff9999',
            edgecolor='gray',
            linewidth=0.3
        )
    else:
        rect = plt.Rectangle(
            (r_flat[i] - d_r / 2, z_flat[i] - d_z / 2),
            d_r, d_z,
            facecolor='lightblue',
            edgecolor='gray',
            linewidth=0.3
        )
    plt.gca().add_patch(rect)

# Overlay cell centres
plt.plot(r_flat, z_flat, 'k.', markersize=0.5, label='Conductor centres')
plt.plot(conductor_center_r, conductor_center_z, 'ro', markersize=0.5)

plt.axhline(0.0, color='#1f77b4', linewidth=0.75, linestyle='-')
plt.axvline(conductor_center_r, color='#1f77b4', linewidth=0.75, linestyle='-')

plt.xlabel('R [m]')
plt.ylabel('Z [m]')

# diagonal line through conductor center
# y = m * x + c
# m = 1.0
# plt.plot(
#     [0.1, 0.7],
#     [0.1, 0.7] - conductor_center_r,
#     color='#1f77b4',
#     linewidth=0.75,
# )


r_line = np.linspace(
    conductor_center_r - 5*d_r,
    conductor_center_r + 5*d_r,
    100
)

z_line = (r_line - conductor_center_r) + conductor_center_z

plt.plot(r_line, z_line, color='#1f77b4', linewidth=0.75)



n_cells = 4.2
plt.xlim(
    conductor_center_r - (n_cells + 0.5) * d_r,
    conductor_center_r + (n_cells + 0.5) * d_r
)
plt.ylim(
    conductor_center_z - (n_cells + 0.5) * d_z,
    conductor_center_z + (n_cells + 0.5) * d_z
)

plt.savefig("grid_for_psi_self_term_investigation.pdf")
plt.show()

```
